```python
year1924 = pd.Series(
    index = ["2019","2020","2021","2022","2023","2024"],
    data = [sub19["역이용자수"].mean(),sub20["역이용자수"].mean(),sub21["역이용자수"].mean(),sub22["역이용자수"].mean()
           ,sub23["역이용자수"].mean(),sub24["역이용자수"].mean()],
    name = "19년부터 24년까지의 역이용자수 평균"
)
```

```python
fig = plt.figure(figsize = (10,5))
plt.plot(year1924, "b-o") # 시리즈의 벨류들이 그대로 y축의 값이 된다, 인덱스는 x축
plt.grid()
plt.xlabel("연도")
plt.ylabel("이용수")
plt.title("19~24년도 역이용자수 평균 추이 (역이용자는 각 역의 승하차인원의 합계)")

plt.show()
```
