```python
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# windows용 한글 폰트 오류 해결
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname = font_path).get_name()
rc("font", family = font_name)
```

```python
total = pd.read_csv("./data/metro/data/1-9호선 통합본(2).csv")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
sns.regplot(x='자전거보관소 대수', y='평균승차총승객수', data=total,
            scatter_kws={'alpha':0.5,'color':'hotpink'}, line_kws={'color':'red'}, ax=axes[0])
axes[0].set_title("자전거보관소 대수와 평균 승차 총 승객수", size =15)
axes[0].set_xlabel("자전거보관소 대수")
axes[0].set_ylabel("평균 승차 총승객수")

sns.regplot(x='자전거보관소 대수', y='평균하차총승객수', data=total,
             scatter_kws={'alpha':0.5}, line_kws={'color':'blue'}, ax=axes[1])
axes[1].set_title("자전거보관소 대수와 평균 하차 총 승객수",size =15)
axes[1].set_xlabel("자전거보관소 대수")
axes[1].set_ylabel("평균 하차 총승객수")

plt.show()

corr_in, p_value_in = stats.pearsonr(total['자전거보관소 대수'], total['평균승차총승객수'])
corr_out, p_value_out = stats.pearsonr(total['자전거보관소 대수'], total['평균하차총승객수'])

print(f"승차,자전거보관소 개수: 상관계수 = {corr_in:.4f}, P-value = {p_value_in:.4f}")
print(f"하차,자전거보관소 개수: 상관계수 = {corr_out:.4f}, P-value = {p_value_out:.4f}")
```
