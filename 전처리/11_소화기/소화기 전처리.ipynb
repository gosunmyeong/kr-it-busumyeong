```python
df["역명"] = df["역명"].str.replace("(1)","")
df["역명"] = df["역명"].str.replace("(2)","")
df["역명"] = df["역명"].str.replace("(3)","")
df["역명"] = df["역명"].str.replace("(4)","")
df["역명"] = df["역명"].str.replace("(5)","")
df["역명"] = df["역명"].str.replace("(6)","")
df["역명"] = df["역명"].str.replace("(7)","")
df["역명"] = df["역명"].str.replace("(8)","")
```
