```python
df = pd.read_csv("./data/서울교통공사_물품보관함 설치수량(1-9호선).csv", encoding= "cp949")
df.head()
```

```python
df["호선"] = df["호선"] + '호선'
```

```python
df = df.drop(df[df["호선"] == "식물원호선"].index)
```

```python
df["보관함명"].unique()
```

```python
df["보관함명"] = df["보관함명"].str.replace("1","")
df["보관함명"] = df["보관함명"].str.replace("2","")
df["보관함명"] = df["보관함명"].str.replace("3","")
df["보관함명"] = df["보관함명"].str.replace("4","")
df["보관함명"] = df["보관함명"].str.replace("5","")
df["보관함명"] = df["보관함명"].str.replace("6","")
df["보관함명"] = df["보관함명"].str.replace("7","")
df["보관함명"] = df["보관함명"].str.replace("8","")
df["보관함명"] = df["보관함명"].str.replace("9","")
df["보관함명"] = df["보관함명"].str.replace("0","")
df["보관함명"] = df["보관함명"].str.replace("~","")
df["보관함명"] = df["보관함명"].str.replace("()","")
df["보관함명"] = df["보관함명"].str.replace("B","")

```

```python
df[df["보관함명"] == "올림픽공원"]
```

```python
df = df.drop(columns = ["연번", "제어부", "열"])
```

```python
df
```

```python
df = df.rename(columns = {"호선" : "노선명", "보관함명" : "역명"})
```

```python
df.head()
```

```python
df["물품보관함 개수"] = df["소형"]+df["중형"]+df["대형"]
```

```python
df.head()
```

```python
locker_count = df.groupby(["노선명","역명"])[["물품보관함 개수"]].sum()
```

```python
locker_count
```

```python
locker_count.reset_index()
```

```python
locker_count.to_csv('1-9호선 역별 보관하 개수.csv', index = False,  encoding = "utf-8-sig")
```
